### What level of acidity receives the highest average rating? low ــــــــــــــــــــــــــــــ

```python
# create a column in the dataframe [acidity_levels] that tells you the levels of the acidity ['high', 'mod_high', 'medium', 'low'] according
# to the level of pH percentiles [min, 25% , 50%, 75%, max] 
# below 25% is high ---  25% to 50% is mod_high [moderately]  --- 50% to 75% medium ---- higher than 75% is low
# There are easier ways to do this than what you think [there is a specific pandas function for this]
```

```python
winequality["pH"].describe()
```

```python
percentiles=winequality["pH"].describe()
def acidity_level(pH_value):
    if pH_value <= percentiles["25%"]:
        return "High"
    elif pH_value <= percentiles["50%"]:
        return "Mod_High"
    elif pH_value <= percentiles["75%"]:
        return "Medium"
    else:
        return "Low"
winequality["acidity_levels"] = winequality["pH"].apply(acidity_level)

```

```python
#another solution (-_-):
percent = [percentiles["min"], percentiles["25%"], percentiles["50%"], percentiles["75%"], percentiles["max"]] 
level = ["High", "Mod_High", "Medium", "Low"] 
winequality["acidity_levels_V2"] = pd.cut(winequality["pH"], percent, labels=level)
winequality.head(10)
```

```python
# Groupby the newely created acidity level column and get the mean of the quality so you can decide which acidity level
mean_acidity_levels=winequality.groupby("acidity_levels")["quality"].mean()
print (mean_acidity_levels)
print("*"*60)
# results in the best quality and answer the question
print("mean quality of each acidity level High : {}".format(mean_acidity_levels[0]))
print("mean quality of each acidity level Low : {}".format(mean_acidity_levels[1]))
print("mean quality of each acidity level Medium : {}".format(mean_acidity_levels[2]))
print("mean quality of each acidity level Mod_High : {}".format(mean_acidity_levels[3]))
```
