# Cleaning Practice
Let's first practice handling missing values and duplicate data using the `cancer_data_means.csv` file.

```python
# import pandas and load cancer data
import pandas as pd
cancer = pd.read_csv("cancer_data_means.csv")

# check which columns have missing values
cancer.isna().any()
#----------OR----------
cancer.columns[cancer.isna().any()]# name of column missing
```

```python
cancer.head(10)
```

```python
print(cancer.shape)
cancer.info()
```

```python
# use the mean to fill in missing values

cancer["texture_mean"]=cancer["texture_mean"].fillna(cancer["texture_mean"].mean())            #full texture_mean missing values to mean
cancer["smoothness_mean"]=cancer["smoothness_mean"].fillna(cancer["smoothness_mean"].mean())   #full smoothness_mean missing values to mean
cancer["symmetry_mean"]=cancer["symmetry_mean"].fillna(cancer["symmetry_mean"].mean())         #full symmetry_mean missing values to mean
#--------OR----------
#cancer.fillna(0)   >>>>>>>>>>>      #full all missing values to zero


# confirm your correction 
cancer.info()
```

```python
# how many duplicates are there ?
cancer.duplicated().sum()
```

```python
# drop duplicates
cancer.drop_duplicates()
```

```python
# confirm correction by rechecking for duplicates in the data
(cancer.drop_duplicates()).duplicated().sum()
```

## Renaming Columns
Since we also previously changed our dataset to only include means of tumor features, the "_mean" at the end of each feature seems unnecessary. It just takes extra time to type in our analysis later. Rename the columns of the dataframe to remove "_mean".

```python
# rename the columns of the dataframe (remove _mean from the name of each column if found)
new=[]
for col in cancer.columns:
    if "_mean" in col:
        new.append(col[:-5])
    else:
        new.append(col)
new

```

```python
# display first few rows of the dataframe to confirm changes
cancer.columns=new
cancer.head(10)
```

```python
# save this for later as a csv file named "cancer_data_edited.csv" and set indexing to false (why ??)
cancer.to_csv("cancer_data_edited.csv",index=False)
```
